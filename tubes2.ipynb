```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
td = pd.read_csv("tennis.csv")
tennis_label = td['playtennis'].replace({'no': 0, 'yes': 1}).values
tennis_data = td.drop(columns=['playtennis'])
# preprocessing string to int
mapping = {'sunny': 0, 'overcast': 1, 'rain': 2, 'hot': 0, 'mild': 1, 'cool': 2, 'high': 0, 'normal'
: 1, True: 1, False: 0}
tennis_data_mapped = tennis_data.replace(mapping).values
print ("Dataset Tennis")
print (tennis_data)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=4, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(tennis_data_mapped, tennis_label, epochs=150, batch_size=1)
# evaluate the model
scores = model.evaluate(tennis_data_mapped, tennis_label)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
